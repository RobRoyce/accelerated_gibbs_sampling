```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
from mpl_toolkits.mplot3d import Axes3D
valid_n = [4 << i for i in range(15)]
valid_k = [2 << i for i in range(7)]
FILE = 'results.csv'
```

```python
df = pd.read_csv(FILE)
df.groupby(by=['N']).describe()
```

```python
df = pd.read_csv(FILE)
df.groupby(by='K').describe()
```

```python
dfs = df.groupby(by=['N', 'K']).mean()
plots = [dfs.loc[1024,:], dfs.loc[2048,:], dfs.loc[4096,:], dfs.loc[8192,:], dfs.loc[16384,:], 
         dfs.loc[32768,:], dfs.loc[65536,:], dfs.loc[131072,:], dfs.loc[262144,:]]    

plt.figure(figsize=(15, 12))

for p in plots:
    plt.subplot()
    plt.plot(p/1000000)

plt.ticklabel_format(style='plain')
plt.grid()
plt.yscale('log')
plt.xscale('linear')
plt.legend(['n=1024', 'n=2048', 'n=4096', 'n=8192', 'n=16384', 'n=32768', 'n=65536', 'n=131072', 'n=262144'])
plt.xlabel('K (classes)')
plt.ylabel('Log Time (seconds)')
plt.xticks([2,4,8,16,32,64,128])
plt.title('GPU(naive) -- Time (sec) vs. K (classes) for fixed N')
```

```python
dfs = df.groupby(by=['K', 'N']).mean()
plots = [dfs.loc[2,:], dfs.loc[4,:], dfs.loc[8,:], dfs.loc[16,:], dfs.loc[32,:], dfs.loc[64,:], dfs.loc[128,:]]    

plt.figure(figsize=(15, 12))

for p in plots:
    plt.subplot()
    plt.plot(p/1000000)

plt.ticklabel_format(style='plain')
plt.grid()
plt.yscale('log')
plt.xscale('log')
plt.legend(['k=2', 'k=4', 'k=8', 'k=16', 'k=32', 'k=64', 'k=128'])
plt.xlabel('N (samples)')
plt.ylabel('Log Time (seconds)')
plt.xticks([1024, 2048, 4096, 8192, 16384, 32768, 65536, 131072, 262144])
plt.title('GPU(naive) -- Time (sec) vs. N (samples) for fixed K')
```
